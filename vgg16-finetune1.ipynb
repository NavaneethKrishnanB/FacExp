```python
train_path = '/kaggle/input/facexp-train/train'
test_path='/kaggle/input/facexpvalidation/validation'
```

```python

from tensorflow.keras.layers import Lambda,Dense,Flatten,Input
from tensorflow.keras.models import Model,Sequential
from tensorflow.keras.applications.vgg16 import VGG16
from tensorflow.keras.preprocessing import image
from tensorflow.keras.preprocessing.image import ImageDataGenerator,load_img
import numpy as np
from glob import glob
```

```python
vgg16 = VGG16(input_tensor=Input(shape=(48,48,3)),weights='imagenet',include_top=False)
```

```python

for layer in vgg16.layers:
    layer.trainable = False
```

```python
folders=glob(train_path+"/*")
print(len(folders))
```

```python
vgg16.summary()
```

```python
from keras.layers import Dropout
```

```python
print(vgg16.layers[-5])
```

```python
newModel = Flatten()(vgg16.layers[-5].output)
newModel=Dense(512,activation='relu')(newModel)
newModel=Dropout(0.2)(newModel)
newModel=Dense(512,activation='relu')(newModel)
newModel=Dropout(0.2)(newModel)
newModel=Dense(len(folders),activation='softmax')(newModel)
```

```python
model = Model(inputs=vgg16.inputs,outputs=newModel)
```

```python
from keras.optimizers import SGD
model.compile(optimizer=SGD(lr=1e-4,momentum=0.9),
             metrics=['accuracy'],
             loss='categorical_crossentropy')
```

```python
from tensorflow.keras.preprocessing.image import ImageDataGenerator
```

```python
train_datagen = ImageDataGenerator(rescale=1./255,
                                  shear_range=0.2,
                                  zoom_range=0.2,
                                  rotation_range=0.2,
                                  horizontal_flip=True,
                                  vertical_flip=True,
                                  validation_split=0.2)
```

```python
test_datagen=ImageDataGenerator(rescale=1./255)
```

```python
train_set = train_datagen.flow_from_directory(
    train_path,
    target_size=(48,48),
    batch_size=32,
    class_mode='categorical',
    shuffle=True,
    subset='training'
)
```

```python
validation_set= train_datagen.flow_from_directory(
    train_path,
    target_size=(48,48),
    batch_size=32,
    class_mode='categorical',
    shuffle=True,
    subset='validation'
)
```

```python
test_set = test_datagen.flow_from_directory(
    test_path,
    target_size=(48,48),
    batch_size=32,
    class_mode='categorical',
    shuffle=False
)
```

```python
from tensorflow.keras.callbacks import ModelCheckpoint
```

```python
checkpoint=ModelCheckpoint('face_exp_vgg16.h5',
                          monitor='val_loss',
                          mode='min',
                          save_best_only=True,
                          verbose=1)
```

```python
from keras.models import load_model
curModel = load_model('../input/vgg16-finetuned2h5/face_exp_vgg16.h5')
```

```python
curModel.summary()
```

```python
import matplotlib.pyplot as plt
```

```python
def fine_tune_layer(model,earlystop,checkpoint):
    model.compile(optimizer=SGD(lr=1e-4,momentum=0.9),
             metrics=['accuracy'],
             loss='categorical_crossentropy')
    history2 = model.fit(x=train_set,
                   validation_data=validation_set,
                   epochs=50,
                   callbacks=[checkpoint,earlystop],
                   steps_per_epoch=len(train_set),
                   validation_steps=len(validation_set))
    plt.plot(history2.history['accuracy'])
    plt.plot(history2.history['val_accuracy'])
    plt.legend(['accuracy','val_accuracy'])
    plt.show()
    plt.plot(history2.history['loss'])
    plt.plot(history2.history['val_loss'])
    plt.legend(['loss','val_los'])
    plt.show()
```

```python
curModel.evaluate_generator(test_set,6158)
```

```python
for layer in curModel.layers:
    print(layer,layer.trainable)
```

```python
from keras.optimizers import SGD
```

```python
earlystop = EarlyStopping(monitor='val_loss',
                          patience = 5,
                        verbose =1,
                        restore_best_weights=True)
checkpoint=ModelCheckpoint('face_exp_vgg16.h5',
                          monitor='val_loss',
                          mode='min',
                          save_best_only=True,
                          verbose=1)
for off in range(14):
    layer = curModel.layers[20-off]

    print(20-off,layer.trainable)
    if layer.trainable:
        continue
    layer.trainable = True
    train_set.reset()
    test_set.reset()
    validation_set.reset()
    fine_tune_layer(curModel,earlystop,checkpoint)
```
